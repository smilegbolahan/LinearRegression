```python
import pandas as pd

company = pd.read_csv("salary_data.csv")
```

```python
company
```

```python
len(company)
```

```python
# Extract the independent and dependent variable
#independent must always be 2D always
#x - independent, y- dependent

#double the square bracket for x, to become 2-Dimensional

independent = company[["YearsExperience"]]
dependent = company["Salary"]
```

```python
independent.ndim

#ndim is used to check the number of dimension. 
#Before predicting our independent variable(column) must be 2D
```

```python
dependent.ndim

#The target must always be 1Dimensional because we have only one outcome
```

# splitting the dataset

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(independent,dependent, test_size=0.2, random_state=0)
```

# call the algorithm we will use on the dataset. (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

# make predictions

```python
model.predict(x_train)
```

```python
table = pd.DataFrame()
#create an empty dataframe and assign to name table


table[0] =  x_train

#create x_train column in table
table[1] =  model.predict(x_train)

#create model predictions on x_train column in table
```

```python
table
```

```python
table2 = pd.DataFrame()
#create an empty dataframe and assign to name table


table2[0] =  x_test

#create x_train column in table
table2[1] =  model.predict(x_test)

#create model predictions on x_train column in table
```

```python
model.predict(x_test)
```

# Evaluate the model

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics


#The mean square error is an important metric to use in 
#evaluating the performance of your machine learning model


#The MSE or mean squared error is always 0 or positive. 

#It returns the average of the sums of the squares of each differences between
#the estimated value and the true value
```

```python
mean_squared_error(y_test,model.predict(x_test))
```

```python
metrics.r2_score(y_test, model.predict(x_test))


#R squared(regression squared) is called the coefficient of determination used to 
#evaluate the performance of the linear regression model

#It is used to check the how well-observed results are reproduced by the model

#best possible score is 1.0
```
